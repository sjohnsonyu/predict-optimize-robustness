```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
missed_df = pd.read_csv('User Id adding Sheet.xlsx - Sheet1.csv')
missed_df['user_id']=missed_df['user_id'].fillna(method='ffill')
missed_df = missed_df[(pd.to_datetime(missed_df['date_time'])>datetime.strptime('2021-04-26', '%Y-%m-%d')) &
          (pd.to_datetime(missed_df['date_time'])<datetime.strptime('2021-06-14', '%Y-%m-%d'))]
```

```python
df = pd.read_csv('outputs/analysis_lists/all_analysis_week_12.csv')

out = pd.merge(df, missed_df, on='user_id')
```
